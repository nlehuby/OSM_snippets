```python
line_template = """

# Ligne {ref}

* réseau / operator : {network}
* couleur de ligne : {color}
* nom de la ligne : {name}
* nombre de trajets : {nb_trip}
""" 


trip_template = """

### Trajet {trip_id}

* {origin_stop_name} -> {destination_stop_name} - {nb_stops} arrêts 
* ligne circulaire : {is_loop}

[voir le trajet]({trip_id}.json) - [charger le trajet dans JOSM](http://localhost:8111/import?new_layer=true&upload_policy=never&url={provider_url}{trip_id}.json)

"""
```

```python
# mapmatching magic

def geojson_to_text(a_geojson):
    flatten_coord_list = [item for sublist in a_geojson['geometry']['coordinates'] for item in sublist]
    shape = []
    for a_couple in a_geojson['geometry']['coordinates']:
        coord = {}
        coord["lon"] = a_couple[0]
        coord["lat"] = a_couple[1]
        shape.append(coord)
    shape_txt = str(shape)
    shape_txt = shape_txt.replace("'","\"")
    return shape_txt



#decode an encoded string
def decode(encoded):
  #six degrees of precision in valhalla
  inv = 1.0 / 1e6;
  decoded = []
  previous = [0,0]
  i = 0
  #for each byte
  while i < len(encoded):
    #for each coord (lat, lon)
    ll = [0,0]
    for j in [0, 1]:
      shift = 0
      byte = 0x20
      #keep decoding bytes until you have this coord
      while byte >= 0x20:
        byte = ord(encoded[i]) - 63
        i += 1
        ll[j] |= (byte & 0x1f) << shift
        shift += 5
      #get the final value adding the previous offset and remember it for the next
      ll[j] = previous[j] + (~(ll[j] >> 1) if ll[j] & 1 else (ll[j] >> 1))
      previous[j] = ll[j]
    #scale by the precision and chop off long coords also flip the positions so
    #its the far more standard lon,lat instead of lat,lon
    #decoded.append([float('%.6f' % (ll[1] * inv)), float('%.6f' % (ll[0] * inv))])
    decoded.append({"lon": float('%.6f' % (ll[1] * inv)), "lat": float('%.6f' % (ll[0] * inv))})
  #hand back the list of coordinates
  return decoded

def from_shape_to_way_list(coord):
    # one only step - trace_attributes
    data = '{"shape":'+ coord +',"costing":"bus","directions_options":{"units":"miles"},"shape_match":"map_snap"}' 
    ta1 = requests.post(trace_attributes_url,  data = data)

    #print(ta1.url)
    #print(ta1.text)
    output = ta1.json()
    if 'error' in output:
        print ('trace_attributes error : {}'.format(output['error']))
        return
    osm_way_list=[]
    for a_way in output['edges']:
        osm_way_list.append('w' + str(a_way['way_id']))
    return osm_way_list 

errors = []
mapmatching_results = {}

if with_mapmatching:
    import os
    import requests
    default_trace_route_url = "https://valhalla.mapzen.com/trace_route?api_key=mapzen-jLrDBSP&"
    default_trace_attributes_url = "https://valhalla.mapzen.com/trace_attributes?api_key=mapzen-jLrDBSP&"

    trace_route_url = os.getenv('valhalla_trace_route_url', default_trace_route_url)
    trace_attributes_url = os.getenv('valhalla_trace_attributes_url', default_trace_route_url)
    
    for elem in trips_full_w_agency.trip_id:  
        file_name = list(trips_full_w_agency[trips_full_w_agency['trip_id']==elem].filename)[0]
        json_name_file = "output/{}.json".format(file_name)
        with open( json_name_file, "r") as json_file:
            try:
                geojson = json.load(json_file)
            except :
                error = {'name': elem, 'error': "invalid json"}
                errors.append(error)
                continue

        
        shape = geojson_to_text(geojson['features'][0])
        tt = from_shape_to_way_list(shape)
        if tt is not None:
            josm_link = "http://localhost:8111/load_object?objects="
            josm_link += ','.join(tt)
            mapmatching_results[elem]=josm_link
            
```

```python
text_result = ""

route_ids = list(set(list(trips_full_w_agency.route_id)))
for route_id in route_ids:
    selected_trips = trips_full_w_agency[trips_full_w_agency['route_id']==route_id]
    elem = selected_trips.iloc[0]
    text_result += line_template.format(
            ref = elem['route_short_name'],
            network = elem['agency_name'],
            color = "#" + elem["route_color"],
            name = elem['route_long_name'],
            nb_trip = len(selected_trips),
        )
    for i, trip in selected_trips.iterrows():
        is_loop = "non"
        if trip['is_loop'] == 1:
            is_loop= "oui"

        text_result += trip_template.format(
            #trip_id = trip['trip_id'],
            trip_id = trip['filename'],
            origin_stop_name = trip['origin_stop_name'],
            destination_stop_name = trip["destination_stop_name"],
            nb_stops = trip['num_stops'],
            is_loop = is_loop,
            provider_url = 'https://raw.githubusercontent.com/nlehuby/jubilant-octo-broccoli/master/{}/'.format(network_name),
        )
        if with_mapmatching:
            if trip['trip_id'] in mapmatching_results:
                text_result += "[charger les chemins du trajet dans JOSM]({})".format(mapmatching_results[trip['trip_id']])
    
text_result
```
